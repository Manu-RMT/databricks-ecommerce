```sql
select * from dim_ecommerce_customers;
select * from dim_ecommerce_products;
select * from fact_ecommerce_order_items;
select * from dim_date limit 5;

CREATE OR REPLACE VIEW v_fact_ecommerce_transaction_all_data AS (
select oi.*,
extract(hour from oi.transaction_ts) as hour_of_day,
p.category_code,
p.category_name,
p.brand_code,
p.brand_name,
p.color,
p.size,
p.material,
p.rating_count,
d.date,
d.year,
d.is_weekend,
d.month,
d.month_name,
d.day_name,
d.quarter
from fact_ecommerce_order_items as oi
left join dim_ecommerce_products p on p.product_id = oi.product_id
left join dim_date d on d.date_key = oi.date_id
)
```
